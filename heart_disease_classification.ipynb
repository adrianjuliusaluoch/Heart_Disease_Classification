### ***Data Preprocessing***

*i) Change object datatype to category since only two groups exist*

```python
cols=[
    'HeartDisease','Smoking','AlcoholDrinking','Stroke','DiffWalking',
    'Sex','Race','Diabetic','PhysicalActivity','GenHealth','Asthma', 'KidneyDisease','SkinCancer']
def change_dtype():
    for index in cols:
        data[index]=data[index].astype('category')
```

```python
change_dtype()
```

*ii) Bin AgeCategory*

*18-24, 25-29, 30-34, 35-39 and 40-44 group into <font color='lightgreen'>Adult</font>
<br/>45-49, 50-54, 55-59, 60-64 and 65-69 group into <font color='lightgreen'>Middle Aged</font>
<br/>70-74, 75-79 and `80 or older` group into <font color='lightgreen'>Elderly</font>*

```python
data['Age']=data['AgeCategory'].apply(
    lambda x : 'Adult' if (x == '18-24'or x == '25-29' or x == '30-34' or x == '35-39' or x == '40-44')
    else 'Middle Aged' if (x == '45-49' or x == '50-54' or x == '55-59' or x == '60-64' or x == '65-69')
    else 'Elderly'
    )
```

### ***Exploratory Data Analysis***

***i) Estimates of Location***

*1) Mean*

```python
data[['BMI','SleepTime']].mean()
```

*2) Trimmed Mean*

```python
sorted_data=data.copy()
```

```python
from scipy.stats import trim_mean
bmi=sorted_data.sort_values(by='BMI',ascending=False).reset_index(drop=True)['BMI']
sleep=sorted_data.sort_values(by='SleepTime',ascending=False).reset_index(drop=True)['SleepTime']
print(f"Trimmed Mean for BMI : {trim_mean(bmi,0.1):,.2f}\nTrimmed Mean for Sleep Time : {trim_mean(sleep,0.1):,.2f}")
```

*3) Weighted Mean*

```python
print(
f"""Weighted Mean BMI on Sleep Time : {np.average(data['BMI'],weights=data['SleepTime']):,.2f}
Weighted Mean Sleep Time on BMI : {np.average(data['SleepTime'],weights=data['BMI']):,.2f}
    """)
```

*4) Median*

```python
data[['BMI','SleepTime']].median()
```

*5) Weighted Median*

```python
import wquantiles as wq
print(
f"""Weighted Median BMI on Sleep Time : {wq.median(data['BMI'],weights=data['SleepTime']):,.2f}
Weighted Median Sleep Time on BMI : {wq.median(data['SleepTime'],weights=data['BMI']):,.2f}
    """)
```
